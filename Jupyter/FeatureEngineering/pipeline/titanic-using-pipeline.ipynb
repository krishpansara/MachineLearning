# Let's Plan

1. ` Filling missing values in Age and embarked `
2. ` Scoring age and Fare `
3. ` OHE on embarked. `
4. ` Selecting best feature ( only for example) `
5. ` Applying model `

   <b> So we create pipeline for this using column transformers <b>

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# imputation transformer

trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')


## output - 2 cols
## shape - (712, 7)
```

```python
# one hot encoding

trf2 = ColumnTransformer([
    # ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,3])

],remainder='passthrough')


## output - 5 cols
## shape - (712, 10)
```

```python
# Scaling

trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])

## output - 10 cols
```

```python
# Feature selection

trf4 = SelectKBest(score_func=chi2,k=5)

## select 8 best cols 
```

```python
# train the model

trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Pipeline Vs make_pipeline

Pipeline requires naming of steps, make_pipeline does not.

(Same applies to ColumnTransformer vs make_column_transformer)

```python
# Alternate Syntax
# pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```
